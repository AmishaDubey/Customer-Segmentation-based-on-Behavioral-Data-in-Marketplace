```python
#imprting libraries
import pandas as pd
df=pd.read_csv('marketing_campaign.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.info()
```

```python
#listed the categories in education
education=df['Education'].unique().tolist()
edu_map={}
df['Education']=le.fit_transform(df['Education'])
edu=df['Education'].unique().tolist()
for i,j in zip(education,edu):
    edu_map[j]=i
print(edu_map)
```

```python
#listed the categories in marital status
marriage=df['Marital_Status'].unique().tolist()
marry_map={}
df['Marital_Status']=le.fit_transform(df['Marital_Status'])
mar=df['Marital_Status'].unique().tolist()
for i,j in zip(marriage,mar):
    marry_map[j]=i
print(marry_map)
```

```python
df.head()
```

```python
#Adding all kind of expenses
df['Expenses']=df['MntWines']+df['MntFruits']+df['MntMeatProducts']+df['MntFishProducts']+df['MntSweetProducts']+df['MntGoldProds']
df.head()
```

```python
df.columns
```

```python
#importing libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#plotting a pie chart for education expenses
edu_exp_map={}
for key in edu_map:
    edu_exp_map[key]=df.loc[df['Education'] == key, 'Expenses'].sum()
edu_count=df['Education'].value_counts().tolist()
edu_count_avg_map={}
for key,val in zip(edu_exp_map,edu_count):
    edu_count_avg_map[key]=edu_exp_map[key]/val
avgg=[]
for key in edu_count_avg_map:
    avgg.append(edu_count_avg_map[key])
avgg_percentage=[]
summ=0
for val in avgg:
    summ+=val
for val in avgg:
    avgg_percentage.append((val/summ)*100)
edu_labels=[]
for key in edu_map:
    edu_labels.append(edu_map[key])
plt.pie(avgg_percentage, labels=edu_labels, colors=['red','green','blue','pink','yellow'], startangle=90, autopct='%1.1f%%')
plt.title('Education Expenses')
plt.show()
```

```python
lll=[]
for i in df['Agee']:
    if i<sum_age_avg:
        lll.append(f"Less than {sum_age_avg}")
    else:
        lll.append(f"More than {sum_age_avg}")
df["Age"]=lll
```
